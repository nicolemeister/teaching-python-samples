Useful functions for printing: Strings are sequences and you can treat them like a list to get individual chars
- .rstrip() = right strip
- .lstrip() = left strip
- splicing string:  "Nicole"[starting_index, ending_index] *not inclusive*
- separation 
- end
- .index to search for the index of a substring: claim.index('c'): finds the index of the c character
- claim.split(): Takes a string and turns it into a list of each separate word
  - You can choose what character to split/join on (ex: ‘-’)
- words.join(): takes list and joins it into a string


```
# .rstrip()
name = "Nicole.txt"
name = name.rstrip('.txt')
print(name)
```

```
# Splicing
# print the first letter
print(name[0])
# print the first letter - 3rd letter
print(name[0:2])
# negative numbers indicate last letter
print(name[-1])
# print everything except the last 2 letters
print(name[:-2])
```

```
# separation 
# default: end a print statment with a new line character
print('hello', 'world', sep=' ')

# these print statements no longer end with a new line character!
print('Mercury', 'Venus', 'Earth', sep=', ', end=', ')
print('Mars', 'Jupiter', 'Saturn', sep=', ', end=', ')

print('Uranus', 'Neptune', 'Pluto', sep=', ')
```

```
name = "Delia"
# returns the index where the e in Delia occurs
name.index('e')
```

```
# split a string by a specific character
names = "Ahmed, Akanshak, Alex, Alex, Arianne, Asmi, Auschwin, Bella, Bhavana, Christian, Delia, Diya, Howard, JJ, Matthew, Melinda, Nicole, Reshmi, Rut, Samuel, Sumaya, Tyler, Tyrell"
name_list = names.split(',')
```

```
joiner = ', '
original_name_string = joiner.join(name_list)
print(original_name_string)
```

## Lists
a. Slicing <br>
b. Useful functions


```
Planets = ["Mercury", "Venus", "Earth", "Mars", "Jupiter", "Saturn", "Uranus", "Neptune"]

# Access element at the end of the list
Planets[-1]
```

```
# Get elements of planets starting from index 0 and not including index 3
Planets[0:3] 
```

```
# Beginning to and not including element at index 3
Planets[:3]
```

```
# From element 3 to the end of the list
Planets[3:]
```

```
# From element 1 up until but not including last one
Planets[1:-1]
```

```
# From 3rd to last element to the end. You can also set the first 3 elements to be something else 
Planets[-3:]
```

## Loops
- if
- elif
- for
- while

```
# if, elif 
planet = "Earth"
if (planet in Planets):
  print(planet + " is a planet")
elif (planet == "Pluto"):
  print(planet + " used to be a planet")
else: print(planet + " is not a planet")
```

```
# FOR LOOPS
for i in range(5):
  print(i)
```

```
# FOR LOOPS
for i,j in enumerate(planets):
  print(i, j)
```

```
# Nested for loops
for i in range(5): 
  for j in range(3):
    print(i, j)
```
