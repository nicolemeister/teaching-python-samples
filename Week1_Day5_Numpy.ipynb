Built In Functions Part 2: Apply a function to the entire array

```
# Sum
print("Sum: ", np.sum(c))

# Product of all elements
print("Product of all the elements: ", np.prod(c))

# Average
print("Average: ", np.mean(c))

# Standard Deviation
print("Standard Deviation: ", np.std(c))

# Variance
print("Variance: ", np.sum(c))

# Min
print("Min: ", np.min(c))

# Index of the array that contains the min
print("Argmin: ", np.argmin(c))

# Max
print("Max: ", np.max(c))

# Index of the array that contains the max
print("Argmax: ", np.argmax(c))
```
